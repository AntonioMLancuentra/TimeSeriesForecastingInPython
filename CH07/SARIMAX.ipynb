```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import pandas as pd
```

```python
df = pd.read_csv('../data/widget_sales.csv')
df.shape
```

```python
window = 2
```

```python
model = SARIMAX(df[:450], order=(0,1,2))
res = model.fit(disp=False)
predictions = res.get_prediction(450, 450 + window - 1)   # changed from (0, i + window - 1) to (i, i + window - 1)
oos_pred = predictions.predicted_mean.iloc[-window:]
oos_pred
```

```python
# Confidence intervals
predictions.conf_int(alpha=0.05)  
# for 95% confidence intervals
```

```python
# Standard error of the predicted mean
predictions.se_mean
```

```python
predictions.summary_frame()
```

```python
# Standard error * Standard error = variance of the predicted mean
predictions.var_pred_mean
```

```python
predictions.dist
# The underlying distribution object, which can be used to calculate custom percentiles or probabilities
```
